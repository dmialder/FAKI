Using libraries

```python
import pandas as pd
import numpy as np
import math
import matplotlib.pyplot as plt
```

Functions and their derivatives

```python
def F_1(x):
    result = math.log(x) + x**2 - 5*x + 5
    return result


def F_2(x):
    result = x**3 - 3*x**2 + 12*x - 12
    return result


def F_1_d(x):
    result = 1/abs(x) + 2*x - 5
    return result


def F_2_d(x):
    result = 3*x**2 - 6*x + 12
    return result


def F_1_dd(x):
    result = -1/x**2 + 2
    return result


def F_2_dd(x):
    result = 6*x - 6
    return result

a1, b1 = 1, 2
a2, b2 = 2, 3.3
a3, b3 = 0.001, 0.1
a4, b4 = 1, 1.5
```

CHARTS OF FUNCTIONS

```python
x1 = np.linspace(-1, 4, 500)
x2 = np.linspace(0, 3, 500)
f01 = 0*x1
f1 = np.log(x1) + x1**2 - 5*x1 + 5
f02 = 0*x2
f2 = x2**3 - 3*x2**2 + 12*x2 - 12


fig, axs = plt.subplots(1, 2)
axs[0].set_title("ln(x)+x^2-5*x+5=0")
axs[0].plot(x1, f1)
axs[0].plot(x1, f01)
axs[0].grid()
axs[1].set_title("x^3-3*x^2+12*x-12=0")
axs[1].plot(x2, f2)
axs[1].plot(x2, f02)
axs[1].grid()
plt.show()
```

DICHOTOMY METHOD

function describrion

```python
def dichotomy_method(a, b, func, eps):
    results = pd.DataFrame(columns=["n", "a", "b", "x", "f(a)", "f(b)", "f(x)"])
    counter = 0
    x = (a+b)/2
    while(abs(func(x)) > eps):
        x = (a + b)/2
        intermediate_results = {"n": counter, "a": a, "b": b, "x": x, "f(a)": func(a), "f(b)": func(b), "f(x)":func(x)}
        if func(x)*func(b) > 0:
            b = x
        else:
            a = x
        results.loc[len(results)] = intermediate_results
        counter += 1
    display(results) 
    return results
```

dichotomy process

```python
eps = 0.0001
dichotomy_results_f1 = []
dichotomy_results_f2 = []

res = dichotomy_method(a1, b1, F_1, eps=eps)
dichotomy_results_f1.append(res["x"][len(res["x"]) - 1])
res = dichotomy_method(a2, b2, F_1, eps=eps)
dichotomy_results_f1.append(res["x"][len(res["x"]) - 1])
res = dichotomy_method(a3, b3, F_1, eps=eps)
dichotomy_results_f1.append(res["x"][len(res["x"]) - 1])
res = dichotomy_method(a4, b4, F_2, eps=eps)
dichotomy_results_f2.append(res["x"][len(res["x"]) - 1])
```

SECANT METHOD

secant describrion

```python
def secant_method(a, b, func, eps):
    results = pd.DataFrame(columns=["n", "a", "b", "f(a)", "f(b)", "x", "f(x)"])
    counter = 0
    x = b
    while(abs(func(x)) > eps):
        intermediate_results = {"n": counter, "a": a, "b": b, "f(a)": func(a), "f(b)": func(b), "x": x, "f(x)":func(x)}
        x = a - ((b - a) * func(a)/(func(b) - func(a)))
        if(func(a)*func(x) > 0):
            a = x
        else:
            b = x
        results.loc[len(results)] = intermediate_results
        counter += 1
    display(results) 
    return results
```

secant process

```python
eps = 0.0001
secant_results_f1 = []
secant_results_f2 = []

res = secant_method(a1, b1, F_1, eps=eps)
secant_results_f1.append(res["x"][len(res["x"]) - 1])
res = secant_method(a2, b2, F_1, eps=eps)
secant_results_f1.append(res["x"][len(res["x"]) - 1])
res = secant_method(a3, b3, F_1, eps=eps)
secant_results_f1.append(res["x"][len(res["x"]) - 1])
res = secant_method(a4, b4, F_2, eps=eps)
secant_results_f2.append(res["x"][len(res["x"]) - 1])
```

NEWTON METHOD

newton function descriprion

```python
def newton_method(x, f, f_d, eps):
    counter = 0
    results = pd.DataFrame(columns=["n", "x", "f(x)", "f'(x)", "-f(x)/f'(x)"])
    while(abs(f(x)) > eps):
        x = x - f(x)/f_d(x)
        intermediate_results = {"n": counter, "x": x, "f(x)": f(x), "f'(x)": f_d(x), "-f(x)/f'(x)":-f(x)/f_d(x)}
        results.loc[len(results)] = intermediate_results
    display(results)
    return results
```

newton function process

```python
eps = 0.0001
newton_results_f1 = []
newton_results_f2 = []

res = newton_method(a1, F_1, F_1_d, eps=eps)
newton_results_f1.append(res["x"][len(res["x"]) - 1])
res = newton_method(b2, F_1, F_1_d, eps=eps)
newton_results_f1.append(res["x"][len(res["x"]) - 1])
res = newton_method(a3, F_1, F_1_d, eps=eps)
newton_results_f1.append(res["x"][len(res["x"]) - 1])
res = newton_method(b4, F_2, F_2_d, eps=eps)
newton_results_f2.append(res["x"][len(res["x"]) - 1])
```

MODED NEWTON METHOD

moded newton method function description

```python
def find_x0(a, b, f, f_dd):
    l = np.linspace(a, b, 21)
    for item in l:
        if f(item)*f_dd(item) > 0:
            return item


def moded_newton_method(a, b, f, f_d, f_dd, eps):
    counter = 0
    
    x = x0 = find_x0(a, b, f, f_dd)
    results = pd.DataFrame(columns=["n", "x0", "x", "f(x)", "f'(x)", "-f(x)/f'(x0)"])
    while(abs(f(x)) > eps):
        x = x - f(x)/f_d(x0)
        intermediate_results = {"n": counter, "x0": x0, "x": x, "f(x)": f(x), "f'(x)": f_d(x), "-f(x)/f'(x0)":-f(x)/f_d(x0)}
        results.loc[len(results)] = intermediate_results
    display(results)
    return results
```

moded nweton method process

```python
eps = 0.0001
moded_newton_results_f1 = []
moded_newton_results_f2 = []

res = moded_newton_method(a1, b1, F_1, F_1_d, F_1_dd, eps=eps)
moded_newton_results_f1.append(res["x"][len(res["x"]) - 1])
res = moded_newton_method(a2, b2, F_1, F_1_d, F_1_dd, eps=eps)
moded_newton_results_f1.append(res["x"][len(res["x"]) - 1])
res = moded_newton_method(a3, b3, F_1, F_1_d, F_1_dd, eps=eps)
moded_newton_results_f1.append(res["x"][len(res["x"]) - 1])
res = moded_newton_method(a4, b4, F_2, F_2_d, F_2_dd, eps=eps)
moded_newton_results_f2.append(res["x"][len(res["x"]) - 1])
```

SIMPLE ITERATION METHOD

simple iteration function description

```python
def phi_1(x):
    result = (math.log(x) + x**2 + 5)/5
    return result


def phi_1_d(x):
    result = (1/abs(x) + 2*x)/5
    return result


def phi_2(x):
    result = -((x**3 - 3*x**2 - 12)/12)
    return result


def phi_2_d(x):
    result = -((x**2 - 2*x)/4)
    return result


def find_x0(a, b, phi_d):
    l = np.linspace(a, b, 101)
    for item in l:
        if abs(phi_d(item)) < 1:
            return item
    
    return b


def simple_iteration_method(a, b, phi, phi_d, eps):
    counter = 0
    results = pd.DataFrame(columns=["n", "x0", "x1", "phi(x0)", "phi(x1)"])
    x0 = find_x0(a, b, phi_d)
    x1 = phi(x0)
    results.loc[len(results)] = {"n": counter, "x0": x0, "x1": x1, "phi(x0)": phi(x0), "phi(x1)": phi(x1)}
    while(abs(x1 - x0) > eps):
        x0 = x1
        x1 = phi(x0)
        intermediate_result = {"n": counter, "x0": x0, "x1": x1, "phi(x0)": phi(x0), "phi(x1)": phi(x1)}
        results.loc[len(results)] = intermediate_result
    display(results)
    return results
```

simple iteration process

```python
eps = 0.0001
simple_iteration_results_f1 = []
simple_iteration_results_f2 = []

res = simple_iteration_method(a1, b1, phi_1, phi_1_d, eps=eps)
simple_iteration_results_f1.append(res["x1"][len(res["x1"]) - 1])
res = simple_iteration_method(a1, b1, phi_1, phi_1_d, eps=eps)
simple_iteration_results_f1.append(res["x1"][len(res["x1"]) - 1])
res = simple_iteration_method(a1, b1, phi_1, phi_1_d, eps=eps)
simple_iteration_results_f1.append(res["x1"][len(res["x1"]) - 1])
res = simple_iteration_method(a4, b4, phi_2, phi_2_d, eps=eps)
simple_iteration_results_f2.append(res["x1"][len(res["x1"]) - 1])
```
